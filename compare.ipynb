```python
from json import dumps
from re import split

from english_words import get_english_words_set
from replace_accents import replace_accents_characters
from requests import get

user_agent = 'grvsmth-biter'
search_suffix = ' in:name'

len_or = len(' OR ')
len_suffix = len(search_suffix)
max_search_length = 256
```

```python
namestext = ''

with open('./names_sorted.txt') as infile:
    namestext = infile.read()

namelines = namestext.split("\n")
namewords = set()

stopwords = get_english_words_set(['web2'], lower=True)

for line in namelines:
    for word in split(r'\s+', line):
        if not word.lower() in stopwords:
            namewords.add(replace_accents_characters(word))
```

```python
search_strings = []
search_string = ''

for index, word in enumerate(namewords):
    if (index + 1) % 6 < 1 or len(search_string) + len_or + len(word) + len_suffix > max_search_length:
        search_strings.append(search_string + search_suffix)
        search_string = word
        continue

    if index > 0:
        search_string += ' OR '

    search_string += word

search_strings.append(search_string + search_suffix)

print(dumps(search_strings))
```

```python
search_url = 'https://api.github.com/search/repositories'

repos = []

for search_string in search_strings:
    print('.')
    res = get(search_url, params={'q': search_string}, headers={'User-Agent': user_agent})
    data = res.json()
    repos += data.get('items')
```

```python
print(len(repos))
```

```python
names = [repo.get('full_name') for repo in repos]
print("\n".join(sorted(names)))
```
